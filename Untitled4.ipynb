```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from scipy.integrate import odeint
import torch
import torch.nn as nn

# Set style for publication-quality plots
plt.style.use('seaborn-v0_8-darkgrid')
plt.rcParams.update({
    'font.family': 'DejaVu Serif',
    'font.size': 12,
    'axes.titlesize': 16,
    'axes.titleweight': 'bold',
    'axes.labelsize': 14,
    'axes.labelweight': 'bold',
    'legend.fontsize': 11,
    'xtick.labelsize': 11,
    'ytick.labelsize': 11,
    'figure.facecolor': 'white',
    'axes.facecolor': '#fafafa',
    'grid.alpha': 0.4,
    'grid.linestyle': '--',
    'lines.linewidth': 2.5,
})

# EXACT parameters from paper
alpha = 0.2
delta1 = 0.08
xi = 0.25
Lambda = 10.2
nu = 0.02
delta2 = 0.09
gamma = 0.01
mu = 9.0
beta_m = 1.5
beta_v = 0.57
beta_h = 0.765343138

# Initial conditions
S0 = 9.0
I0 = 0.1

# Time domain
t_span = np.linspace(0, 200, 2000)

# EXACT ODE system from paper (reduced SI model)
def ode_system(y, t):
    S, I = y
    dS = Lambda - (beta_h * S * I) / (1 + alpha * I) - (delta1 + nu) * S + (1 - beta_v) * delta2 * I
    dI = (beta_h * S * I) / (1 + alpha * I) + beta_v * delta2 * I - (delta2 + gamma) * I - (beta_m * I) / (1 + mu * I)
    return [dS, dI]

# STEP 1: Classical ODE solution
print("ðŸ”„ Solving classical ODE...")
y0 = [S0, I0]
solution = odeint(ode_system, y0, t_span)
S_classical = solution[:, 0]
I_classical = solution[:, 1]

# STEP 2: Add 5% Gaussian noise to I(t)
np.random.seed(42)
noise_level = 0.05
I_noisy = I_classical + noise_level * np.std(I_classical) * np.random.randn(len(I_classical))

# Prepare PyTorch tensors
t_tensor = torch.tensor(t_span.reshape(-1, 1), dtype=torch.float32, requires_grad=True)
I_data = torch.tensor(I_noisy.reshape(-1, 1), dtype=torch.float32)
S0_tensor = torch.tensor([[S0]], dtype=torch.float32)
I0_tensor = torch.tensor([[I0]], dtype=torch.float32)

# STEP 3: PINN Implementation
class PINN(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(1, 64),
            nn.Tanh(),
            nn.Linear(64, 64),
            nn.Tanh(),
            nn.Linear(64, 64),
            nn.Tanh(),
            nn.Linear(64, 64),
            nn.Tanh(),
            nn.Linear(64, 2)
        )

    def forward(self, t):
        out = self.net(t)
        S = torch.sigmoid(out[:, 0:1]) * 15.0
        I = torch.sigmoid(out[:, 1:2]) * 1000.0
        return S, I

model = PINN()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, factor=0.5, patience=500)

# Training
print("\nðŸš€ Training PINN...")
epochs = 15000
t_colloc = t_tensor.clone().detach().requires_grad_(True)

loss_history = []
data_loss_history = []
physics_loss_history = []
ic_loss_history = []

for epoch in range(epochs):
    optimizer.zero_grad()

    # Predictions
    S_pred, I_pred = model(t_colloc)

    # Compute derivatives
    S_t = torch.autograd.grad(S_pred.sum(), t_colloc, create_graph=True)[0]
    I_t = torch.autograd.grad(I_pred.sum(), t_colloc, create_graph=True)[0]

    # Physics equations
    horiz_trans = (beta_h * S_pred * I_pred) / (1 + alpha * I_pred)
    treatment = (beta_m * I_pred) / (1 + mu * I_pred)
    vertical = beta_v * delta2 * I_pred

    # Residuals
    residual_S = S_t - (Lambda - horiz_trans - (delta1 + nu) * S_pred + (1 - beta_v) * delta2 * I_pred)
    residual_I = I_t - (horiz_trans + vertical - (delta2 + gamma) * I_pred - treatment)

    # Initial conditions
    t0 = torch.tensor([[0.0]], requires_grad=True)
    S0_pred, I0_pred = model(t0)

    # Data loss
    S_data_pred, I_data_pred = model(t_tensor)
    loss_data = torch.mean((I_data_pred - I_data)**2)

    # Physics loss
    loss_physics = torch.mean(residual_S**2) + torch.mean(residual_I**2)

    # Initial condition loss
    loss_ic = (S0_pred - S0_tensor)**2 + (I0_pred - I0_tensor)**2

    # Total loss
    loss = 1000.0 * loss_data + 10.0 * loss_physics + 500.0 * loss_ic

    loss.backward()
    optimizer.step()
    scheduler.step(loss)

    # Store losses
    loss_history.append(loss.item())
    data_loss_history.append(loss_data.item())
    physics_loss_history.append(loss_physics.item())
    ic_loss_history.append(loss_ic.item())

    if (epoch + 1) % 1500 == 0:
        print(f"Epoch {epoch+1:5d}/{epochs} | Loss: {loss.item():.4e} | Data: {loss_data.item():.4e} | Physics: {loss_physics.item():.4e}")

# STEP 4: Predictions
print("\nðŸ“Š Generating predictions...")
with torch.no_grad():
    S_pinn, I_pinn = model(t_tensor)
    S_pinn = S_pinn.numpy().flatten()
    I_pinn = I_pinn.numpy().flatten()

# Compute errors
rel_error_S = np.linalg.norm(S_pinn - S_classical) / np.linalg.norm(S_classical)
rel_error_I = np.linalg.norm(I_pinn - I_classical) / np.linalg.norm(I_classical)
rmse_S = np.sqrt(np.mean((S_pinn - S_classical)**2))
rmse_I = np.sqrt(np.mean((I_pinn - I_classical)**2))
mae_S = np.mean(np.abs(S_pinn - S_classical))
mae_I = np.mean(np.abs(I_pinn - I_classical))

print(f"\n{'='*70}")
print("ðŸ“ˆ RESULTS:")
print(f"{'='*70}")
print(f"  Infected Population I(t):")
print(f"    â€¢ Relative L2 Error: {rel_error_I:.6f} ({rel_error_I*100:.3f}%)")
print(f"    â€¢ RMSE: {rmse_I:.6f}")
print(f"    â€¢ MAE:  {mae_I:.6f}")
print(f"\n  Susceptible Population S(t):")
print(f"    â€¢ Relative L2 Error: {rel_error_S:.6f} ({rel_error_S*100:.3f}%)")
print(f"    â€¢ RMSE: {rmse_S:.6f}")
print(f"    â€¢ MAE:  {mae_S:.6f}")
print(f"{'='*70}\n")

# ============================================================================
# BEAUTIFUL PLOTS - Publication Quality
# ============================================================================

# Create comprehensive figure
fig = plt.figure(figsize=(20, 12))
gs = GridSpec(3, 3, figure=fig, hspace=0.35, wspace=0.35,
              left=0.06, right=0.97, top=0.94, bottom=0.06)

# Color palette (professional colors)
color_classical = '#1f77b4'  # Blue
color_pinn = '#d62728'       # Red
color_data = '#7f7f7f'       # Gray
color_error = '#ff7f0e'      # Orange
color_loss = '#9467bd'       # Purple

# ============================================================================
# Plot 1: Infected Population (Large, main focus)
# ============================================================================
ax1 = fig.add_subplot(gs[0:2, 0:2])
ax1.scatter(t_span[::5], I_noisy[::5], c=color_data, s=8, alpha=0.35,
            label='Noisy Training Data (5% noise)', zorder=1, marker='o', edgecolors='none')
ax1.plot(t_span, I_classical, color=color_classical, linewidth=3.5,
         label='Classical ODE (Ground Truth)', zorder=3, linestyle='-')
ax1.plot(t_span, I_pinn, color=color_pinn, linewidth=2.8,
         label='PINN Prediction', zorder=4, linestyle='--', alpha=0.9)

ax1.set_xlabel('Time (days)', fontweight='bold', fontsize=15)
ax1.set_ylabel('Infected Population I(t)', fontweight='bold', fontsize=15)
ax1.set_title('Infected Population: PINN vs Classical ODE Solution',
              fontsize=17, fontweight='bold', pad=20)
ax1.legend(loc='upper left', frameon=True, fancybox=True, shadow=True,
           fontsize=12, framealpha=0.95)
ax1.grid(True, alpha=0.4, linestyle='--', linewidth=0.8)
ax1.set_xlim([0, 200])

# Add error box
textstr = (f'Relative L2 Error: {rel_error_I:.4f}\n'
           f'RMSE: {rmse_I:.4f}\n'
           f'MAE: {mae_I:.4f}')
props = dict(boxstyle='round,pad=0.7', facecolor='#fff3cd', alpha=0.9,
             edgecolor='#856404', linewidth=2.5)
ax1.text(0.72, 0.30, textstr, transform=ax1.transAxes, fontsize=12,
         verticalalignment='top', bbox=props, family='monospace', weight='bold')

# ============================================================================
# Plot 2: Susceptible Population
# ============================================================================
ax2 = fig.add_subplot(gs[0:2, 2])
ax2.plot(t_span, S_classical, color=color_classical, linewidth=3.5,
         label='Classical ODE', zorder=2, linestyle='-')
ax2.plot(t_span, S_pinn, color=color_pinn, linewidth=2.8,
         label='PINN Prediction', zorder=3, linestyle='--', alpha=0.9)

ax2.set_xlabel('Time (days)', fontweight='bold', fontsize=14)
ax2.set_ylabel('Susceptible S(t)', fontweight='bold', fontsize=14)
ax2.set_title('Susceptible Population\nComparison', fontsize=15, fontweight='bold', pad=15)
ax2.legend(loc='best', frameon=True, fancybox=True, shadow=True,
           fontsize=11, framealpha=0.95)
ax2.grid(True, alpha=0.4, linestyle='--', linewidth=0.8)
ax2.set_xlim([0, 200])

# Add error box
textstr = f'Rel. L2: {rel_error_S:.4f}\nRMSE: {rmse_S:.4f}'
props = dict(boxstyle='round,pad=0.6', facecolor='#d1ecf1', alpha=0.9,
             edgecolor='#0c5460', linewidth=2.5)
ax2.text(0.05, 0.95, textstr, transform=ax2.transAxes, fontsize=11,
         verticalalignment='top', bbox=props, family='monospace', weight='bold')

# ============================================================================
# Plot 3: Phase Portrait (S vs I)
# ============================================================================
ax3 = fig.add_subplot(gs[2, 0])
ax3.plot(S_classical, I_classical, color=color_classical, linewidth=3,
         label='Classical ODE', alpha=0.9)
ax3.plot(S_pinn, I_pinn, color=color_pinn, linewidth=2.5,
         label='PINN', linestyle='--', alpha=0.85)
ax3.scatter([S0], [I0], s=200, c='green', marker='*',
            edgecolors='darkgreen', linewidths=2, zorder=5, label='Initial State')
ax3.scatter([S_classical[-1]], [I_classical[-1]], s=150, c='navy',
            marker='s', edgecolors='black', linewidths=2, zorder=5, label='Final State')

ax3.set_xlabel('Susceptible S(t)', fontweight='bold', fontsize=13)
ax3.set_ylabel('Infected I(t)', fontweight='bold', fontsize=13)
ax3.set_title('Phase Portrait', fontsize=14, fontweight='bold', pad=12)
ax3.legend(loc='upper right', frameon=True, fancybox=True, shadow=True, fontsize=10)
ax3.grid(True, alpha=0.4, linestyle='--')

# ============================================================================
# Plot 4: Training Loss (Log Scale)
# ============================================================================
ax4 = fig.add_subplot(gs[2, 1])
epochs_plot = np.arange(1, len(loss_history) + 1)
ax4.semilogy(epochs_plot, loss_history, color=color_loss, linewidth=2.5, label='Total Loss', alpha=0.9)
ax4.semilogy(epochs_plot, data_loss_history, color=color_classical, linewidth=1.8,
             label='Data Loss', alpha=0.7, linestyle='-.')
ax4.semilogy(epochs_plot, physics_loss_history, color=color_pinn, linewidth=1.8,
             label='Physics Loss', alpha=0.7, linestyle=':')

ax4.set_xlabel('Epochs', fontweight='bold', fontsize=13)
ax4.set_ylabel('Loss (log scale)', fontweight='bold', fontsize=13)
ax4.set_title('Training Convergence', fontsize=14, fontweight='bold', pad=12)
ax4.legend(loc='best', frameon=True, fancybox=True, shadow=True, fontsize=10)
ax4.grid(True, alpha=0.4, linestyle='--', which='both')

# ============================================================================
# Plot 5: Prediction Error Over Time
# ============================================================================
ax5 = fig.add_subplot(gs[2, 2])
error_I = I_pinn - I_classical
error_S = S_pinn - S_classical

ax5.plot(t_span, error_I, color=color_pinn, linewidth=2.5, label='Error I(t)', alpha=0.8)
ax5.plot(t_span, error_S, color=color_error, linewidth=2.5, label='Error S(t)', alpha=0.8)
ax5.axhline(y=0, color='black', linestyle='-', linewidth=1.5, alpha=0.6)
ax5.fill_between(t_span, 0, error_I, color=color_pinn, alpha=0.2)
ax5.fill_between(t_span, 0, error_S, color=color_error, alpha=0.2)

ax5.set_xlabel('Time (days)', fontweight='bold', fontsize=13)
ax5.set_ylabel('Prediction Error', fontweight='bold', fontsize=13)
ax5.set_title('PINN Error Distribution', fontsize=14, fontweight='bold', pad=12)
ax5.legend(loc='best', frameon=True, fancybox=True, shadow=True, fontsize=10)
ax5.grid(True, alpha=0.4, linestyle='--')
ax5.set_xlim([0, 200])

# Main title
fig.suptitle('Physics-Informed Neural Network (PINN) for Monkeypox Model\n' +
             'Reduced SI Model with Saturated Incidence & Treatment (Equation 3)',
             fontsize=19, fontweight='bold', y=0.98)

# Save high-resolution figure
plt.savefig('pinn_monkeypox_publication_quality.png', dpi=400, bbox_inches='tight',
            facecolor='white', edgecolor='none')
print("âœ… Saved: 'pinn_monkeypox_publication_quality.png'")

# ============================================================================
# BONUS: Side-by-side comparison (cleaner version)
# ============================================================================
fig2, axes = plt.subplots(1, 2, figsize=(18, 7))

# Infected
axes[0].scatter(t_span[::5], I_noisy[::5], c=color_data, s=10, alpha=0.35,
                label='Noisy Data (5%)', zorder=1)
axes[0].plot(t_span, I_classical, color=color_classical, linewidth=3.5,
             label='Classical ODE', zorder=2)
axes[0].plot(t_span, I_pinn, color=color_pinn, linewidth=2.8,
             label='PINN', linestyle='--', zorder=3, alpha=0.9)
axes[0].set_xlabel('Time (days)', fontweight='bold', fontsize=15)
axes[0].set_ylabel('Infected Population I(t)', fontweight='bold', fontsize=15)
axes[0].set_title('Infected Population Comparison', fontsize=17, fontweight='bold', pad=15)
axes[0].legend(fontsize=13, frameon=True, shadow=True, fancybox=True)
axes[0].grid(alpha=0.4, linestyle='--')
textstr = f'Rel. L2 Error: {rel_error_I:.4f}\nRMSE: {rmse_I:.4f}'
props = dict(boxstyle='round', facecolor='wheat', alpha=0.85, edgecolor='#856404', linewidth=2)
axes[0].text(0.70, 0.30, textstr, transform=axes[0].transAxes, fontsize=13,
             bbox=props, family='monospace', weight='bold')

# Susceptible
axes[1].plot(t_span, S_classical, color=color_classical, linewidth=3.5,
             label='Classical ODE', zorder=2)
axes[1].plot(t_span, S_pinn, color=color_pinn, linewidth=2.8,
             label='PINN', linestyle='--', zorder=3, alpha=0.9)
axes[1].set_xlabel('Time (days)', fontweight='bold', fontsize=15)
axes[1].set_ylabel('Susceptible Population S(t)', fontweight='bold', fontsize=15)
axes[1].set_title('Susceptible Population Comparison', fontsize=17, fontweight='bold', pad=15)
axes[1].legend(fontsize=13, frameon=True, shadow=True, fancybox=True)
axes[1].grid(alpha=0.4, linestyle='--')
textstr = f'Rel. L2 Error: {rel_error_S:.4f}\nRMSE: {rmse_S:.4f}'
props = dict(boxstyle='round', facecolor='lightblue', alpha=0.85, edgecolor='#0c5460', linewidth=2)
axes[1].text(0.05, 0.95, textstr, transform=axes[1].transAxes, fontsize=13,
             bbox=props, family='monospace', weight='bold')

plt.tight_layout()
plt.savefig('pinn_monkeypox_sidebyside_beautiful.png', dpi=400, bbox_inches='tight',
            facecolor='white')
print("âœ… Saved: 'pinn_monkeypox_sidebyside_beautiful.png'")

plt.show()

print("\n" + "="*70)
print("ðŸŽ‰ ALL DONE! Beautiful plots generated successfully!")
print("="*70)
```
